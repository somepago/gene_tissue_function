```python
# def correlation_dict_function(expression_dict):
genes = gene_exp_dict.keys()
corr_dict = {}
gene_combinations = list(itertools.combinations(genes, 2))
for i in gene_combinations:
    gene1,gene2 = i
    corr, p-val = spearmanr(*i)
    if p-val<0.05:
        corr_dict[i] = corr
        corr_dict[(gene2,gene1)] = corr
    else:
        print(i)
        corr_dict[i] = ''
        corr_dict[(gene2,gene1)] = ''
```

```python
genes = gene_exp_dict.keys()
corr_dict = {}
gene_combinations = list(itertools.combinations(genes, 2))
count =0
for i in gene_combinations:
    gene1,gene2 = i
    corr, p_val = spearmanr(gene_exp_dict[gene1],gene_exp_dict[gene2])
    if p_val<0.05:
        corr_dict[i] = corr
        corr_dict[(gene2,gene1)] = corr
    else:
        print(i)
        count+=1
```

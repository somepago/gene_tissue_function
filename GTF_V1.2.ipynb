```python
gtex_shortlisted_w_genes = pd.read_csv('gtex_shortlisted_w_genes.tsv',sep='\t')
dtg_masterlist = pd.read_csv('Disease_Tissue_Gene_masterlist_updated.csv')
dtg_masterlist.head()
```

```python
tissue_mapping = pd.read_csv('dtpaper-gtextissue-mapping.csv')
tissue_mapping.head()
tissue_mapping_dict = {key for key in set(tissue_mapping[' Disease-tissue-paper'])}
tt_mapping_dict = {}

for index, row in tissue_mapping.iterrows():
    key  = row[' Disease-tissue-paper']
    val = row['Gtex-highlevel-tissue']
    tt_mapping_dict[key] = val
    
```

```python
dtg_masterlist['Primary_Tissue'] = dtg_masterlist['Tissue'].map(tt_mapping_dict)
```

```python
dtg_masterlist.head()
```

```python
tg_map = {k: list(v) for k, v in dtg_masterlist.groupby('Primary_Tissue')['Gene Symbols']}

for key in tg_map.keys():
    appended_list = []
    vals = tg_map[key]
    for i in vals:
        tokens = i.split(',')
        for j in tokens:
            appended_list.append(j.strip())
    tg_map[key] = list(set(appended_list))
    
```

```python
tg_map['Adipose Tissue']
```

```python
shortlisted_genes = list(gtex_shortlisted_w_genes['Gene'])
gt_pairs_frompaper = []
for key in tg_map.keys():
    vals = tg_map[key]
    for i in vals:
        if i in shortlisted_genes:
            gt_pairs_frompaper.append(str(i+'_'+key))
```
